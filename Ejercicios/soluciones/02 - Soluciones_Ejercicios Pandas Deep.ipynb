## Ejercicio 8

Crea una función lambda para poner en minúscula la columna `continent`

```python
def go_lower(x):
    return x.lower()
```

```python
go_lower('AS')
```

```python
go_lower = lambda x: x.lower()
```

```python
go_lower('AS')
```

```python
df=df.dropna()
```

```python
df.continent.apply(go_lower)
```
