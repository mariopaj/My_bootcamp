# Ejercicio 3

Crea un dataframe con todos los personajes.

```python
import pandas as pd
```

```python
req.get(url).json()
```

```python
%%time

people=[]

for i in range(1, 827):
    
    data=req.get(f'https://rickandmortyapi.com/api/character/{i}').json()
    
    people.append(data)
    
    
df=pd.json_normalize(people)

df.shape
```

```python
%%time

people=[]

for i in range(1, 43):
    
    data=req.get(f'https://rickandmortyapi.com/api/character/?page={i}').json()['results']
    
    people+=data
    
    
df=pd.json_normalize(people)

df.shape
```

```python
%%time

list_char=[i+1 for i in range(826)]

res_all=req.get(f"https://rickandmortyapi.com/api/character/{list_char}")

df2=pd.DataFrame(res_all.json())

df2.shape
```
