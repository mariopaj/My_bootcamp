## Limpieza de valoren nulos

```python
len(df)
```

```python
# Eliminamos filas a partir de la 6302 (todas sonfilas nulas a partir de esta)

df=df.drop(range(6302,len(df)), axis=0)

df.shape
```

```python
df.tail()
```

```python
df.info(memory_usage="deep")
```

```python
nan_cols=df.isna().sum()

nan_cols[nan_cols>0]
```

```python
df[df.year==0]
```

```python
# Eliminamos las filas en las que no tengamos el año 0.0 lo contamos como NaN

df.drop(df[df.year==0].index, inplace = True)

df.tail()
```

```python
# Eliminamos las filas que tienen menos de 600 NaN

df = df.dropna(subset = nan_cols[nan_cols<600].index)
```

```python
nan_cols=df.isna().sum()

nan_cols[nan_cols>0]
```

```python
df.shape
```

```python
# Rellenamos las columnas unnammed con unknow y species

df[["unnamed: 22","unnamed: 23","species"]]=df[["unnamed: 22","unnamed: 23","species"]].fillna("unknow")
```

```python
nan_cols=df.isna().sum()

nan_cols[nan_cols>0]
```

```python
# Las variables age y time son objetos por lo que no podemos rellenar los Nan con su media
# eliminamos las filas que tengan valores nulos de estas columnas

df = df.dropna(subset=["age","time"])
```

```python
df.shape
```

```python
# Vamos a convertir la columna age en int

df["age"] = pd.to_numeric(df["age"], errors = 'coerce')

df.info()
```

```python
df.reset_index(inplace=True, drop=True)
```
