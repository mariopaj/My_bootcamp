```python
# cargar datos

import pandas as pd

df=pd.read_csv('../data/shoes.csv')

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
# seleccion de datos

df=df[['brand', 'colors', 'sizes', 'prices.amountMax', 'imageURLs', 'ean']]

df.head()
```

```python
df['imageURLs'][0]
```

```python
# quita nan

df=df.dropna().reset_index(drop=True)

df.head()
```

```python
df.info()
```

```python
df.sizes=df.sizes.apply(lambda x: x.split(','))
df.colors=df.colors.apply(lambda x: x.split(','))

df.head()
```

```python
# un registro por talla y color

df=df.explode('sizes').reset_index(drop=True)
df=df.explode('colors').reset_index(drop=True)

df.head()
```

```python
df.info()
```

### Inserta datos

```python
# insert query, insert into nombre_tabla (columnas, nombre...) values 1,2,3..;


for i in df.index:
    
    insert_query='insert into {} ({}) values {};'.format('zapatos', 
                                                         ','.join(df.columns), 
                                                         tuple(df.iloc[i].values))
    
    cursor.execute(insert_query)

db.commit()  # esto machaca la db, aseguramos el guardado
```

```python
df.head()
```

```python
# insert query, insert into nombre_tabla (columnas, nombre...) values 1,2,3..;

tabla='zapatos'

columnas=','.join(df.columns)  # devuelve una string con los nombres separados por comas


for i in df.index:  # para cada fila del df....
    
    datos=tuple(df.iloc[i].values) # convierte los datos a tupla...
    
    insert_query=f'insert into {tabla} ({columnas}) values {datos};'  # string de python como query de sql 
    
    cursor.execute(insert_query) # darle al rayo

db.commit()  # esto machaca la db, aseguramos el guardado
```

```python
# check

cursor.execute('select * from zapatos where price>47 limit 1')

for x in cursor:
    print(x)
```

```python
str_conn='mysql+pymysql://root:password@localhost:3306/publications'

motor=create_engine(str_conn)
```

```python
query='''
        select group_concat(t.title) as titles,
        t.year
        from (select title, year(max(pubdate)) as 'year',
        count(titleauthor.title_id) as num_titles,
        group_concat(concat(authors.au_fname, ' ', authors.au_lname)) as authors
        from titles
        left join titleauthor
        on titles.title_id=titleauthor.title_id
        left join authors
        on authors.au_id=titleauthor.au_id
        group by titles.title
        order by num_titles desc) as t
        group by t.year;

'''
```

```python
df=pd.read_sql(query, motor)

df.head()
```
