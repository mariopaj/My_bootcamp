# Obtener datos.

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
raw_data=pd.read_csv('train.csv')

raw_data.head()
```

```python
raw_data.info()
```

```python
X_raw=raw_data.drop('Churn', axis=1)

y=raw_data.Churn

X_raw.head()
```

```python
X_numeric=X_raw._get_numeric_data().drop('id', axis=1)

X_numeric.head()
```

```python
# categoricas importantes: Partner, InternetService, Contract, PaymentMethod

X_categorico=pd.DataFrame()

    
X_categorico['Partner']=X_raw.Partner.map(lambda x: 1 if x=='Yes' else 0)

X_categorico[['DSL', 'Fiber optic', 'No']]=pd.get_dummies(X_raw.InternetService)

X_categorico[['Month-to-month', 'Two year', 'One year']]=pd.get_dummies(X_raw.Contract)

X_categorico[['Electronic check', 
              'Bank transfer (automatic)',
              'Credit card (automatic)', 
              'Mailed check']]     =   pd.get_dummies(X_raw.PaymentMethod)


X_categorico.head()
```

```python
X=pd.concat([X_numeric, X_categorico], axis=1)

X.head()
```

```python
def transform_data(df):
    
    df_numeric=df._get_numeric_data().drop('id', axis=1)
    
    df_categorico=pd.DataFrame()

    
    df_categorico['Partner']=df.Partner.map(lambda x: 1 if x=='Yes' else 0)

    df_categorico[['DSL', 'Fiber optic', 'No']]=pd.get_dummies(df.InternetService)

    df_categorico[['Month-to-month', 'Two year', 'One year']]=pd.get_dummies(df.Contract)

    df_categorico[['Electronic check', 
                  'Bank transfer (automatic)',
                  'Credit card (automatic)', 
                  'Mailed check']]     =   pd.get_dummies(df.PaymentMethod)
    
    X=pd.concat([df_numeric, df_categorico], axis=1)
    
    return X
```

```python
X=transform_data(X_raw)

X.head()
```

```python
y.head()
```

# Modelado

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# modelos

from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.ensemble import GradientBoostingClassifier as GBC
from xgboost import XGBClassifier as XGBC
from lightgbm import LGBMClassifier as LGBMC

from sklearn.metrics import f1_score
```

```python
def fitter(modelo):
    
    modelo.fit(X_train, y_train)
    
    y_pred=modelo.predict(X_test)
    
    score=f1_score(y_test, y_pred)
    
    print('F1:', score, modelo)
    
    return modelo
```

```python
modelos=[RFC(), GBC(), XGBC(), LGBMC()]

for m in modelos:
    fitter(m)
```

```python
# nuestro modelo ahora va a ser GBC, hacemos gridsearching

modelo=GBC()
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV # random mas rapido
```

```python
params={'loss': ['deviance', 'exponential'],
        'learning_rate': [0.01, 0.05, 0.1, 0.15, 0.2],
        'n_estimators': [50, 100, 150, 500]}
```

```python
grid=GridSearchCV(modelo, params, cv=2)

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
modelo=fitter(grid.best_estimator_)
```

### Ya tenemos nuestro "mejor" modelo. Ahora se reentrena con todo.

```python
modelo=modelo.fit(X, y)
```
